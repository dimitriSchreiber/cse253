# Computational Graph in Numpy

Define the inputs, tagets and parameters

```python
import numpy as np

X = np.random.rand(10, 4)
W1 = np.random.rand(4, 8)
W2 = np.random.rand(8, 5)
# Assume targets as one-hot encoded labels
targets = np.random.rand(10, 5)
```

Forward Pass

```python
def sigmoid(x):
    return 1.0/(1.0+np.exp(-x))

A1 = np.dot(X, W1)
Z1 = sigmoid(A1)
A2 = np.dot(Z1, W2)
y = np.exp(A2)/np.sum(np.exp(A2), axis = 1, keepdims = True)
neg_loss = np.sum(targets * np.log(y))
```

Backward Pass

```python
delta2 = targets - y
dW2 = np.dot( np.transpose(Z1), delta2)
dZ1 = np.dot(delta2, np.transpose(W2))
delta1 = sigmoid(A1) * (1 - sigmoid(A1)) * dZ1
dW1 = np.dot(np.transpose(X), delta1)
dX = np.dot( delta1, np.transpose(W1))
```

# Computational Graphs in PyTorch

Define input and parameter variables

```python
import torch
from torch.autograd import Variable

dtype = torch.FloatTensor
X = Variable( torch.randn(10, 4).type(dtype) , requires_grad=False)
W1 = Variable( torch.randn(4, 8).type(dtype) , requires_grad=True)
W2 = Variable( torch.randn(8, 5).type(dtype) , requires_grad=True)
# Assume targets as one-hot encoded labels
targets = Variable( torch.randn(10, 5) , requires_grad=False)
```

Forward Pass (Very Similar to Numpy)

```python
A1 = torch.mm(X, W1)
Z1 = torch.sigmoid(A1)
A2 = torch.mm(Z1, W2)
y = torch.exp(A2)/torch.sum(torch.exp(A2), dim = 1).unsqueeze(1)
neg_loss = torch.sum(targets * torch.log(y))
```

Backward Pass (Just this!)


```python
neg_loss.backward()
```

```python
print W1.grad
print W2.grad
```

Weight Update

```python
learning_rate = 0.001
W1.data += learning_rate * W1.grad.data
W2.data += learning_rate * W2.grad.data
```

## Custom Autograd Functions in PyTorch

```python
class Relu(torch.autograd.Function):
    def forward(self, x):
        self.save_for_backward(x)
        return x.clamp(min = 0)

    def backward(self, grad_y):
        x, = self.saved_tensors
        grad_input = grad_y.clone()
        grad_input[x < 0] = 0
        return grad_input
```

Same Computational Graph as above. ReLU instead of Sigmoid

```python
dtype = torch.FloatTensor

X = Variable( torch.randn(10, 4).type(dtype) , requires_grad=False)
W1 = Variable( torch.randn(4, 8).type(dtype) , requires_grad=True)
W2 = Variable( torch.randn(8, 5).type(dtype) , requires_grad=True)
# Assume targets as one-hot encoded labels
targets = Variable( torch.randn(10, 5) , requires_grad=False)

# FORWARD PASS
A1 = torch.mm(X, W1)
relu = Relu()
Z1 = relu(A1)
A2 = torch.mm(Z1, W2)
y = torch.exp(A2)/torch.sum(torch.exp(A2), dim = 1).unsqueeze(1)
neg_loss = torch.sum(targets * torch.log(y))

# BACKPROP : THAT'S IT!!!
neg_loss.backward()

learning_rate = 0.001
W1.data += learning_rate * W1.grad.data
W2.data += learning_rate * W2.grad.data
```

## Same Computational Graph - Higher Level Abstraction : nn

Setting up data. Spiral data instead of random. Source : http://cs231n.github.io/neural-networks-2/

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
%load_ext autoreload
%autoreload 2
# for auto-reloading extenrnal modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
np.random.seed(0)
N = 100 # number of points per class
D = 2 # dimensionality
K = 3 # number of classes
X = np.zeros((N*K,D)).astype('float32')
targets = np.zeros(N*K, dtype='int64')
for j in xrange(K):
  ix = range(N*j,N*(j+1))
  r = np.linspace(0.0,1,N) # radius
  t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
  X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
  targets[ix] = j
fig = plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=targets, s=40, cmap=plt.cm.Spectral)
plt.xlim([-1,1])
plt.ylim([-1,1])
```

```python
import torch
from torch.autograd import Variable
import relu_pytorch
import numpy as np

dtype = torch.FloatTensor


torch_X = Variable( torch.from_numpy(X), requires_grad=False)
# Random Targets : Note not one-hot but actual labels
torch_targets = Variable( torch.from_numpy(targets), requires_grad=False)

model = torch.nn.Sequential(
            torch.nn.Linear(D, 30),
            torch.nn.ReLU(),
            torch.nn.Linear(30, K)
        )

# CrossEntropyLoss combines softmax and the loss functions
loss_fn = torch.nn.CrossEntropyLoss()
learning_rate = 0.01

optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

for i in range(500):
    A2 = model(torch_X)
    loss = loss_fn(A2, torch_targets)

    model.zero_grad()
    loss.backward()

    optimizer.step()

    print loss.data
```

```python
h = 0.1
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
predict_on_X = np.c_[xx.ravel(), yy.ravel()].astype('float32')
print "predict_on_X", predict_on_X
Z = model(Variable( torch.from_numpy(predict_on_X))).data.numpy()
print "check Z", Z
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

fig = plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=targets, s=40, cmap=plt.cm.Spectral)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```

## Same Computsational Graph : Using nn.Module

Define Module : Only the setup and forward pass

```python
class MyNet(torch.nn.Module):
    def __init__(self):
        super(MyNet, self).__init__()
        self.linear1 = torch.nn.Linear(D, 30)
        self.linear2 = torch.nn.Linear(30, K)
        self.activation = torch.nn.ReLU()

    def forward(self, x):
        a1 = self.linear1(x) 
        z1 = self.activation(a1)
        a2 = self.linear2(z1)

        return a2
```

```python
dtype = torch.FloatTensor

torch_X = Variable( torch.from_numpy(X), requires_grad=False)
# Random Targets : Note not one-hot but actual labels
torch_targets = Variable( torch.from_numpy(targets), requires_grad=False)


model = MyNet()

# CrossEntropyLoss combines softmax and the loss functions
loss_fn = torch.nn.CrossEntropyLoss()
learning_rate = 0.01

optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)

for i in range(300):
    A2 = model(torch_X)
    loss = loss_fn(A2, torch_targets)

    model.zero_grad()
    loss.backward()

    optimizer.step()

    print loss
```

```python
h = 0.1
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
predict_on_X = np.c_[xx.ravel(), yy.ravel()].astype('float32')
print "predict_on_X", predict_on_X
Z = model(Variable( torch.from_numpy(predict_on_X))).data.numpy()
print "check Z", Z
Z = np.argmax(Z, axis=1)
Z = Z.reshape(xx.shape)

fig = plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=targets, s=40, cmap=plt.cm.Spectral)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
